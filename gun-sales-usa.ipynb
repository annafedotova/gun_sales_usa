## Introduction

This project explores the trends and tendencies of gun sales in the USA in the past 20 years. As a proxy for gun sales the FBI's National Criminal Background Checks System (NICS) information about background checks performed through the system is being used.

### Main dataset:

- **FBI NICS Firearm Background Check Data**

This dataset comes from the following [Github repository](https://github.com/BuzzFeedNews/nics-firearm-background-checks) and is based on the FBI's National Instant Criminal Background Check System. The NICS is used to determine whether a prospective buyer is eligible to buy firearms or explosives. Gun shops call into this system to ensure that each customer does not have a criminal record or isn’t otherwise ineligible to make a purchase.The FBI provides data on the number of firearm checks by month, state, and type, and currently covers the period of November 1998 – July 2018.

### Additional datasets:

- **State level data from census.gov**

This dataset includes census data for 2010 and 2016 per state with population related metrics, such as race, gender, education level, employment, etc.

- **USA 2016 Presidential Election by County**

This dataset contains the results of the 2016 US presidential elections as proportion of votes in favor of Republicans and Democrats per county. It comes from [OpenDataSoft](https://data.opendatasoft.com/explore/dataset/usa-2016-presidential-election-by-county%40public/table/).

### Important caveats of NICS dataset

The Github repository provides some important notes and caveats about the dataset:

> These statistics represent the number of firearm background checks initiated through the NICS. They do not represent the number of firearms sold. Based on varying state laws and purchase scenarios, a one-to-one correlation cannot be made between a firearm background check and a firearm sale.

A bit more background, from The Trace:

> The FBI’s background check numbers come with caveats: As seen in the late February-early March 2014 bubble, many checks are for concealed carry permits, not actual gun sales. Kentucky runs a new check on each concealed carry license holder each month. And of course, the FBI’s numbers don’t include private gun sales, many of which do not require a background check. A forthcoming study conducted by Harvard researchers found that roughly 40 percent of respondents had acquired their most recent firearm without going through a background check. Despite those vagaries, the FBI’s NICS numbers are widely accepted as the best proxy for total gun sales in a given time period.

Not all categories of background checks may be equally useful/pertinent to your research. When The New York Times analyzed NICS data in Dec. 2015, it included this methodological note:

> Note: Sales estimates are calculated from handgun, long gun and multiple-gun background checks. Permit checks and other categories of background checks are excluded. In California, multiple-gun checks were excluded because data was inconsistent. Because state laws differ, sales levels between states cannot be directly compared.

The authors of that NYT analysis describe how they used the NICS data to estimate gun sales:

> To convert background checks into estimated sales, we relied on a method suggested in the Small Arms Survey by Jurgen Brauer, a professor at Georgia Regents University. Each long gun and handgun check was counted as 1.1 sales. Each multiple-gun check was counted as two sales. Permit checks and other types of checks were omitted. The multiplier is an estimate based on Mr. Brauer's interviews with gun shop owners.

## Data Wrangling

In this section we will focus on understadning and cleaning the data from the three datasets and finally creating a combined dataset that will help us explore the relationships between census and elections data per state and the gun sales data per state.


### NICS dataset

```python
# loading the gun dataset
df = pd.read_csv('gun_data.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

Most of the datatypes seem correct, but in some columns there are significant amounts of null values.

First, let's convert the month column into datetime format and create a separate column with year

```python
# converting month from string to datetime and adding a separate column for year
df['month'] = df['month'].apply(lambda x: pd.to_datetime(x))
df['year'] = df['month'].apply(lambda x: str(x.year))
```

Second, let's see which metrics can be used as a proxy for gun sales

```python
# checking the top states based on mean monthly totals
df.groupby('state').totals.mean().astype(int).sort_values(ascending=False)
```

Based on mean totals, Kentucky is the top state by far. However, this big difference compared to the second top state (Texas) seems too large. Let's look more in detail where this difference come from.

```python
df.query('state == "Kentucky"').head()
```

Looks like almost 400,000 permit checks were completed per month in Kentucky in the most recent months (around 10% of its whole population).

```python
# checking the proportion of permit checks in totals
state = df.groupby('state').mean()
state['permit_share'] = (100*(state['permit']) / state['totals']).round(2)

state[['permit_share', 'totals']].sort_values(by='permit_share',ascending=False)
```

Out of average total monthly checks in Kentucky 85% correspond to permit checks. After doing some [research online](https://wfpl.org/kentucky-background-checks-stand-out/), turns out that
> Kentucky runs a new check on each concealed carry license holder each month.

This means that the number of permit checks does not actually reflect the background checks related to gun sales.
The description of the dataset on [Github repository](https://github.com/BuzzFeedNews/nics-firearm-background-checks), also indicates that leaving out the permit checks might be a better approach.

> Not all categories of background checks may be equally useful/pertinent to your research. When The New York Times analyzed NICS data in Dec. 2015, it included this methodological note:

>> Note: Sales estimates are calculated from handgun, long gun and multiple-gun background checks. Permit checks and other categories of background checks are excluded. 


We will leave out the permit checks and focus only on the checks related to specific type of gun sales

```python
# checking the missing data for the metrics that we are interested in
df[['handgun', 'long_gun', 'other', 'multiple', 'private_sale_handgun','private_sale_long_gun', 'private_sale_other']].info()
```

Looks like most of the missing values appear in private sales data. Let's look more in detail at the private sales metric.

```python
# replacing null values with zeros
df = df.fillna(0)
```

```python
df['public_sales'] = df['handgun'] + df['long_gun'] + df['other'] + df['multiple']
df['private_sales'] = df['private_sale_handgun'] + df['private_sale_long_gun'] + df['private_sale_other']

df.groupby('state')[['private_sales', 'public_sales']].sum().sort_values(by='private_sales', ascending=False)
```

```python
df.groupby('year')['private_sales'].sum().sort_values(ascending=False)
```

It seems that the data on private sales checks only appears from 2013 on and in some states there is no registers of private sales at all (even though there are background checks related to public sales in these states). 

After checking the [federal regulations on background checks](https://lawcenter.giffords.org/gun-laws/policy-areas/background-checks/universal-background-checks/), it is clear that

> Although federal law requires licensed firearms dealers to perform background checks on prospective purchasers, it does not require unlicensed sellers to do so. 

It is up to each state whether they will oblige private sales to go through the background check system.

Therefore for the purpose of this analysis we will focus only on checks related to public sales.

```python
# droping the columns that will not be used
df = df[['month', 'state', 'public_sales', 'handgun', 'long_gun', 'year']]

df.head()
```

### Census dataset

```python
cens = pd.read_csv('U.S. Census Data.csv')
cens.drop(['Fact Note'], axis=1, inplace=True)

cens
```

```python
# removing unnecessary columns and rows
cens = cens.iloc[[0,3,6,7,8,10,12,13,14,15,18,19,20,21,34,35,48,51,62]]
cens.set_index('Fact',inplace=True)
cens = cens.transpose()

cens.head()
```

```python
# renaming the columns
cens.columns = ['population_16', 'population_10', 'under_18_16', 'under_18_10', 'over_65', 'female', 'white', 'black', 'indian',\
'asian', 'hisp', 'white_non_hisp', 'veteran', 'foreign', 'high_school', 'bachelor', 'income', 'employed', 'density']

cens.head()
```

```python
# checking the data types 
cens.info()
```

```python
# fixing the data types
def clean_cens_data(data):
    for column in data.keys():
        data[column] = data[column].str.replace('$','')
        data[column] = data[column].str.replace(',','')
        data[column] = data[column].apply(lambda x: float(x.replace('%',''))/100 if '%' in x else x)
        data[column] = data[column].apply(pd.to_numeric)
    return data
        
cens = clean_cens_data(cens)

cens.head()
```

```python
# confirming that all the columns have correct data types after the cleaning
cens.info()
```

```python
# adding extra metrics - employment rate and proportion of veterans  
cens['employment_rate'] = (cens['employed'] / (cens['population_16'] * (1 - cens['under_18_16'] - cens['over_65']))).round(4)
cens['veteran_rate'] = (cens['veteran'] / cens['population_16']).round(4)
```

### Elections dataset

```python
elect = pd.read_csv('usa-2016-presidential-election-by-county.csv', sep=';')
elect.head()
```

```python
# filtering out the columns we will need
elect = elect[['State','Votes','Republicans 2016', 'Democrats 2016']]

elect.head()
```

```python
# adding a metric for number of voters for each party
elect['votes_rep'] = elect['Votes']*elect['Republicans 2016']/100
elect['votes_dem'] = elect['Votes']*elect['Democrats 2016']/100
```

```python
# aggregating the voters for each party by state
elect.rename(columns={'Votes': 'votes'}, inplace=True)
elect = elect.groupby('State')[['votes','votes_rep','votes_dem']].sum()

elect.head()
```

```python
# dropping the state with NA values - Alaska
elect = elect.dropna()
```

```python
# adding the proportion of voters for each party by state
elect['rep'] = elect['votes_rep'] / elect['votes']
elect['dem'] = elect['votes_dem'] / elect['votes']

elect.head()
```

### Combining the datasets

The combined dataset will be used for exploring the relationships between census and elections data per state and the amount of background checks.
Since both the census and the election datasets have data for 2016, we will take the background checks for 2016 as a proxy for gun sales for this year.

```python
# creating a table with sales per state for 2010 and 2016
sales_16 = df[df['year'] == '2016']
sales_10 = df[df['year'] == '2010']

sales_by_state_16 = sales_16.groupby('state').public_sales.sum()
sales_by_state_10 = sales_10.groupby('state').public_sales.sum()
```

```python
# merging census data with the amount of public sales related background checks
cmb = pd.merge(sales_by_state_16.to_frame(), cens, left_index=True, right_index=True)
cmb.rename(columns={'public_sales': 'public_sales_16'}, inplace=True)

cmb = pd.merge(sales_by_state_10.to_frame(), cmb, left_index=True, right_index=True)
cmb.rename(columns={'public_sales': 'public_sales_10'}, inplace=True)

cmb.head()
```

```python
# calculating sales per 1000 adults in 2010 and 2016, taking into account only people older than 18
cmb['sales_per_mille_16'] = 1000 * cmb['public_sales_16'] / (cmb['population_16'] * (1 - cmb['under_18_16']))
cmb['sales_per_mille_10'] = 1000 * cmb['public_sales_10'] / (cmb['population_10'] * (1 - cmb['under_18_10']))
```

```python
# merging elections dataset with the combined dataset
cmb = elect.merge(cmb, right_index=True, left_index=True)

cmb.head()
```

### Q1: What are the trends and tendencies of gun sales over time?

```python
# grouping the public sales data by month
monthly = df.groupby('month').sum()
monthly.head()
```

```python
# plotting the monthly amount of background checks related to public gun sales over time
sns.set_style("darkgrid")

plt.xlabel('Year')
plt.ylabel('Background Checks')
plt.title('Evolution of background checks for public sales, 1998 - 2018')
monthly['public_sales'].plot(kind='line', figsize=(16,5))

plt.show()
```

The time series plot reveals the presence of seasonality in the data, with big peaks of background checks in the end of each year (November - December). These peaks can be related to the holiday season sales starting with Black Friday sales in the end of November.

```python
# decomposing the time series into trend, seasonal and residual components
decompose = seasonal_decompose(monthly['public_sales'], model='multiplicative')

plt.rcParams["figure.figsize"] = (20,10)
decompose.plot()
plt.show()
```

The trend component shows a clear upward trend in background checks over the past 20 years. Let's smooth the data using moving averages to see any peaks in the data that are not related to the seasonality.

```python
# smoothing the time series using moving averages
monthly['sales_ma'] = monthly['public_sales'].rolling(window=12, center=True).mean()

plt.xlabel('Year')
plt.ylabel('Background Checks')
plt.title('Evolution of background checks for public sales, seasonally adjusted, 1999 - 2018')
monthly['sales_ma'].plot(figsize=(16,5))

plt.show()
```

Looking at the evolution of background checks adjusted for seasonality, there are some peaks in data that are not related to the seasonal component:

- **End of 2008**: coincides with Obama first election
- **End of 2012**: coincides with Obama second election and the Sandy Hook Elementary School shooting 
- **End of 2015 - Beginning of 2016**: San Bernardino and Orlando mass shootings

According to a [New York Times article](https://www.nytimes.com/interactive/2015/12/10/us/gun-sales-terrorism-obama-restrictions.html) that analysed trends in gun sales, there is:

> a pattern of spikes in sales after terrorist attacks and calls for stricter gun-buying laws

### Q2: What is the split of sales by gun type?

```python
# calculating a moving average by gun type
monthly['handgun_ma'] = monthly['handgun'].rolling(window=12, center=True).mean()
monthly['long_gun_ma'] = monthly['long_gun'].rolling(window=12, center=True).mean()

monthly[['handgun_ma', 'long_gun_ma']].plot(kind='line', figsize=(16,5))
plt.xlabel('Year')
plt.ylabel('Background Checks')
plt.title('Evolution of background checks by gun type, seasonally adjusted, 1999 - 2018')
plt.legend(('Handgun', 'Long Gun'), loc='upper left', fontsize=12, frameon=True, facecolor='white')

plt.show()
```

We can see from the graph that up until 2014, long guns were more in demand than hanguns. However, in 2014 the checks for handguns overcome the checks for long guns. 

As for the spikes in demand for guns, they tend to affect both handguns and long guns similarly. 
Altough if we compare the spikes of the end of 2008 and 2012, with the most recent spike of the end of 2015 - beginning of 2016, it seems that the latter brought bigger increase in hangun sales compared to long gun sales.

Handguns are more associated with self-defense, and since the latest spike occured after several mass shootings, this could be a factor that would explain bigger surge of the demans for handguns.

```python
# calculating a proportion of sales by gun type
monthly['handgun_share'] = (monthly['handgun'] / monthly['public_sales']).rolling(window=12, center=True).mean()
monthly['long_gun_share'] = (monthly['long_gun'] / monthly['public_sales']).rolling(window=12, center=True).mean()

monthly[['handgun_share', 'long_gun_share']].plot(kind='line', figsize=(16,5))
plt.xlabel('Year')
plt.ylabel('Share of Gun Type')
plt.title('Evolution of background checks by gun type, seasonally adjusted, 1999 - 2018')
plt.legend(('Handgun', 'Long Gun'), loc='upper right', fontsize=12, frameon=True, facecolor='white')

plt.show()
```

The evolution of background checks by gun type shows that while long guns constituted a larger proportion of sales in the past (up to 68% in 2003), they were overcome by handguns in 2014. In 2017 handguns approximately constituted 55% of all background checks, whereas long guns stayed at around 40% of all checks).

```python
# calculationg the split by gun type per state
by_state = df.groupby('state').mean()
by_state = by_state.sort_values(by='public_sales', ascending=True)

by_state['handgun_share'] = (by_state['handgun'] / by_state['public_sales'])
by_state['long_gun_share'] = (by_state['long_gun'] / by_state['public_sales'])
```

```python
# finding states with the majority of checks related to handguns
top_handgun = by_state[by_state['handgun_share'] > 0.5].sort_values(by='handgun_share', ascending=True)

fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.set_title('Split of Checks by Gun Type in States with Preference for Handguns, %', fontsize=14)
ax1.xaxis.set_label_text('Proportion of Checks by Gun Type')

top_handgun[['handgun_share', 'long_gun_share']].plot(kind='barh', stacked=True, ax=ax1, legend=None)

ax2 = fig.add_subplot(122)
ax2.set_title('Average Number of Checks by Gun Type in States with Preference for Handguns', fontsize=14)
ax2.xaxis.set_label_text('Average Monthly Checks')
ax2.legend(loc='upper right', fontsize=14, frameon=True, facecolor='white')

top_handgun[['handgun', 'long_gun']].plot(kind='barh', stacked=True, ax=ax2)

plt.legend(('Handgun', 'Long Gun'), loc='upper right', fontsize=12, frameon=True, facecolor='white')

plt.show()
```

From the graph we can see that top 3 states with stronger preference for handguns over long guns have very low amount of monthly background checks (less than 1,000 per month). The top 3 states with highest amount of checks per month out of the states that have preference to handguns over long guns are: 

- Florida
- Tennessee
- Illinois

```python
# finding states with the majority of checks related to long guns
top_long_gun = by_state[by_state['long_gun_share'] > 0.5].sort_values(by='long_gun_share', ascending=True)

fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.set_title('Split of Checks by Gun Type in States with Preference for Long Guns, %', fontsize=14)
ax1.xaxis.set_label_text('Proportion of Checks by Gun Type')

top_long_gun[['long_gun_share', 'handgun_share']].plot(kind='barh', stacked=True, ax=ax1, legend=None, color=\
                                                            (sns.color_palette("deep")[1], sns.color_palette("deep")[0]))

ax2 = fig.add_subplot(122)
ax2.set_title('Average Number of Checks by Gun Type in States with Preference for Long Guns', fontsize=14)
ax2.xaxis.set_label_text('Average Monthly Checks')
ax2.legend(('Long Gun', 'Handgun'), loc='upper right', fontsize=14, frameon=True, facecolor='white')

top_long_gun[['long_gun', 'handgun']].plot(kind='barh', stacked=True, ax=ax2, color=\
                                                            (sns.color_palette("deep")[1], sns.color_palette("deep")[0]))

plt.legend(('Long Gun', 'Handgun'), loc='upper right', fontsize=12, frameon=True, facecolor='white')

plt.show()
```

Overall, there are more states with preference for long guns, but we must take into account that although historically long guns constituted larger proportion of sales, since 2014 they were overcome by the handguns, so it is possible that in the upcoming years more states will have a higher preference for handguns.



The top 3 states with highest amount of checks per month out of the states that have preference to handguns over long guns are:

- Texas
- Pennsylvania
- Missouri

### Q3: Which states have had the highest growth in gun registrations?

```python
# comparing amount of background checks in 2010 vs 2016 per state
cmb[['sales_per_mille_10', 'sales_per_mille_16']].sort_values(by='sales_per_mille_16', ascending=False).plot(kind='bar')
plt.xlabel('State')
plt.ylabel('Amount of Background Checks per Thousand Adults')
plt.title('Background Checks per Thousand Adults, 2010 vs 2016', fontsize=14)
plt.legend(('2010', '2016'), loc='upper right', fontsize=14, frameon=True, facecolor='white')

plt.show()
```

All of the states, except for Nebraska and Iowa, have experienced an increase in background checks per adult from 2010 to 2016.

```python
# calculating growth rate of sales per thousand adults from 2010 to 2016
cmb['spm_abs_growth'] = cmb['sales_per_mille_16'] - cmb['sales_per_mille_10']
cmb['spm_growth_rate'] = 100*(cmb['sales_per_mille_16'] - cmb['sales_per_mille_10'])/cmb['sales_per_mille_10']
```

```python
# finding states with highest increase in background checks from 2010 to 2016
fig = plt.figure()

ax1 = fig.add_subplot(121)
ax1.set_title('Top 10 States with Highest Absolute Growth in Background Checks, 2010-2016', fontsize=14)
ax1.xaxis.set_label_text('Absolute Increase in Background Checks per Thousand Adults')

cmb['spm_abs_growth'].sort_values(ascending=True).iloc[-10:].plot(kind='barh', ax=ax1)

ax2 = fig.add_subplot(122)
ax2.set_title('Top 10 States with Highest Growth Rate in Background Checks, 2010-2016', fontsize=14)
ax2.xaxis.set_label_text('Growth Rate of Background Checks per Thousand Adults, %')

cmb['spm_growth_rate'].dropna().sort_values(ascending=True).iloc[-10:].plot(kind='barh', ax=ax2)

plt.show()
```

Here is the list of top 10 states with highest increase of background checks per adult, in terms of absolute increase and in terms of growth rate. In some states the amount of background checks more than doubled. 

The states with the highest absolute increase in background checks per capita are: 

- New Hampshire
- Indiana 
- South Dakota

The states with the highest growth rate in background checks per capita are:

- Delaware
- California
- New Jersey

The following states appeared in top 10 highest increase of background checks per adult, both in terms of absolute increase and in terms of growth rate:

- New Hampshire
- Indiana
- Delaware

- Almost all of the states have experienced an increase in background checks per capita between 2010 and 2016
- In some states the amount of background checks per capita more than doubled between 2010 and 2016 
- The states with the highest absolute increase in background checks per capita are: New Hampshire, Indiana and South Dakota.

**4- What census data is most associated with high level of background checks per capita**

- share of votes for Republican party
- proportion of white non-hispanic population
- proportion of veterans

The following census data appears to be positively correlated with the level of background checks: 

- share of votes for Democratic party
- proportion of asian population
